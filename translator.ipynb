```python
!pip install deep-translator --quiet

from deep_translator import GoogleTranslator

# نمایش زبان‌ها
def show_languages():
    print("\nAvailable language codes (examples):")
    print("en = English")
    print("fa = Persian")
    print("ar = Arabic")
    print("fr = French")
    print("de = German")
    print("es = Spanish")
    print("ru = Russian")
    print("hi = Hindi")
    print("zh-CN = Chinese (Simplified)")
    print("-" * 40)

# تابع ترجمه متن
def translate_text():
    text = input("\nEnter text to translate: ").strip()
    if not text:
        print("Error: Text cannot be empty.")
        return

    target = input("Translate to (language code, e.g. en, fr, fa): ").strip()

    try:
        translated = GoogleTranslator(source='auto', target=target).translate(text)
        print(f"\nTranslated text → {translated}\n")
    except Exception as e:
        print("Translation failed. Please check the language code.")

# منوی اصلی
def main():
    print("===== Python Translator =====")

    while True:
        print("\nChoose an option:")
        print("1. Show language codes")
        print("2. Translate text")
        print("3. Exit")

        choice = input("Enter choice (1/2/3): ")

        if choice == "1":
            show_languages()
        elif choice == "2":
            translate_text()
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
